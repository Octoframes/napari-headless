```python
import napari
from napari.plugins.io import read_data_with_plugins
from pathlib import Path
import numpy as np
viewer = napari.Viewer()
path = str(Path.cwd() /  "Knee")
data = read_data_with_plugins([path]) 

stack = data[0][0][0]



viewer.add_image(stack, colormap="gray_r") 
stack
```

```python
# Python

# Calculate the middle indices
middle_x = stack.shape[0] // 2
middle_y = stack.shape[1] // 2
middle_z = stack.shape[2] // 2

# Define the size of the crop around the middle
crop_size = 50  # Adjust this value as needed

# Calculate the start and end indices for the crop
start_x = middle_x - crop_size
end_x = middle_x + crop_size
start_y = middle_y - crop_size
end_y = middle_y + crop_size
start_z = middle_z - crop_size
end_z = middle_z + crop_size

# Crop the stack around the middle
stack2 = stack[start_x:end_x, start_y:end_y, start_z:end_z]
stack2
```

```python
viewer.add_image(stack2, colormap="gray_r") 
```
